```python
import os
import pandas as pd
```

```python
path = "data/company_reports"
file_list = []
for files in os.listdir(path):
    file_list.append(files)
print(file_list)
```

```python
split_file_list = [x.replace('.csv', '').split("_") for x in file_list]
split_file_list
```

```python
name_list = [x[1] for x in split_file_list]
name_list
```

```python
name_list = list(set(name_list))
name_list.sort()
name_list
```
